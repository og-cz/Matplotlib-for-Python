# Error Bar
An error bar is not the data point.It is a visual representation of uncertainty, variation, or possible range for that data point.

**Think of it like saying:**

*“This value is 10, but it could reasonably be between 9 and 12.”*

Error bars are graphical representations of the variability of data. They are used to indicate the error or uncertainty in a reported measurement. Error bars give a general idea of how precise a measurement is. For example, say we want to give a prediction about car sales for the next 12 months, but we are not 100% sure about our prediction. In order to indicate this uncertainty, we would provide a relative possible error

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

##### but here the paramters are:
- X = must be labels which can be letter/s or numeric
- Y = must be numeric value
- yerr (Y-error) = must be number within the rnage of Y which shows error of that bar

```python
# bar plot
fig, axe = plt.subplots()

yerr = [2, 3, 10, 4] # new data for errors
axe.errorbar(label1, data1, yerr=yerr) # its just like normal line plot but with error which is yerr
plt.show()
```

```python
# draw multiple errorbar in a plot
fig, axe = plt.subplots()

yerr1 = [2, 3, 10, 4]
yerr2 = [1, 1.5, 5, 2]

axe.errorbar(label1, data1, yerr=yerr1) 
axe.errorbar(label1, data2, yerr=yerr2) # we used label1 to just show comparison
plt.show()
```

```python
# draw plot with 2 figure-plot as one figure
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(10,3)) # 10x3 size

yerr1 = [2, 3, 10, 4]
yerr2 = [1, 1.5, 5, 2]

axe[0].errorbar(label1, data1, yerr1) # first col
axe[1].errorbar(label1,data2, yerr2) # second col
plt.show()
```
