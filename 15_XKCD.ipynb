# XCKD

XKCD = A famous webcomic
- Created by Randall Munroe, known for stick-figure drawings and nerd humor.

Matplotlib includes a built-in drawing style that makes your plots look like they were hand-drawn, similar to the XKCD comic style.

It makes the plot look sketchy, wiggly, and funny.

##### you can turn ANY Matplotlib plot into XKCD style USING:
```python
with plt.xckd():
    # code your plots here
```

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

````python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
````

````python
# import the needed libs
import matplotlib.pyplot as plt
````

````python
with plt.xkcd():
    fig, axe = plt.subplots()
    axe.plot(data1,label1)
    axe.text(25,'a','"welcome to the xkcd world\nthis is the xkcd style"')
    plt.show()
````
