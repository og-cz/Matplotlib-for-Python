# Elements inside a plot

Below are the key terms that describe what exists inside a plot. These help you understand the structure of a figure and how each part can be customized.

- Figure - the entire drawing area or canvas that holds everything.
- Axes - the actual plotting area where data is drawn (the “graph” inside the figure).
- Axis (x-axis / y-axis) - lines that define coordinate directions for the data.
- Ticks - small marks along the axes showing scale values.
- Tick Labels - the numbers/text beside each tick.
- Spines - the border lines around the plotting area.
- Title - text displayed above the plot.
- Axis Labels - text labeling the x-axis and y-axis.
- Data Lines / Markers - the actual plotted data (lines, dots, etc.).
- Legend - explains what each line or marker in the plot represents.
- Grid - optional horizontal/vertical reference lines behind the data.
- Annotations - text or arrows highlighting specific points.
- Fill Areas - shaded regions such as confidence bands or highlighted zones.
- Images / Shapes / Patches - optional elements like rectangles, circles, etc.
- Color Map - a color gradient applied to heatmaps or scatter plots.
- Subplots - multiple Axes inside one Figure.

### Figure vs Axes
A Figure can contain:
- titles
- multiple plots
- images
- text
- decorations
- multiple Axes

Each Axes contains:
- x-axis and y-axis
- ticks
- spines (borders)
- grid lines
- legend
- labels
- data lines, markers, bars, etc.
- annotations
