```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# fill area 
fig, axe = plt.subplots()
axe.plot(data1, data2) # here is the line
axe.fill_between(data1, data2, facecolor='g', alpha=0.6) # and it filled the area 
plt.show()
```

```python
# fill a plot with different colors
fig, axe = plt.subplots()
axe.plot(data1, data2)
axe.fill_between(data1, data2, facecolor='g', alpha=0.6)

data2_scaled = [y * 1.2 for y in data2] # scaled data2

axe.fill_between(data1, data2_scaled, facecolor='b', alpha=0.3)
plt.show()
```

# How to draw a confidence band
A confidence band always needs ONLY two things:

1. An estimate (a curve or line).

Example:

- the mean
- a fitted line (regression)
- a smoothed curve
- a prediction from a model

2. The uncertainty of that estimate.

Example:

- standard deviation
- standard error
- regression error
- bootstrap variation

That’s all.

```python
error = [5,5,5,5] # band
x = data1

lower = [data1[_] - error[_] for _ in range(len(error))]
upper = [data1[_] + error[_] for _ in range(len(error))]

x = range(len(data1))

fig, axe = plt.subplots()
axe.plot(x, data1, '-o', label='estimate')
axe.fill_between(x, lower, upper, alpha=0.3, label='bruteforce band')
axe.legend()
plt.show()
```
