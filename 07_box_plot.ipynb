# Box plot
A box plot (also called a box-and-whisker plot) is a statistical chart that shows the distribution of your data using five summary numbers.

A box plot breaks your data into 5 key values:

- Minimum -> The smallest non-outlier value in the data. It shows where the lower end of your normal data begins.
- Q1 (25th percentile) -> 25% of the data is below this value. Marks the start of the lower half of the distribution.
- Median (50th percentile) -> The middle value of the data. Half the numbers are below it, half are above.
- Q3 (75th percentile)  -> 75% of the data is below this value. Marks the start of the upper half of the distribution.
- Maximum -> The largest non-outlier value in the data. It shows where the upper end of your normal data ends.

It also highlights outliers.

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# Box plot
fix, axe = plt.subplots()
axe.boxplot(data1, labels=label1[0]) # it only need one label and the list of data of that label
plt.show()
```

```python
# draw multiple boxplot in a plot
fig, axe = plt.subplots()
axe.boxplot([data1, data2], labels=[label1[0], label2[0]]) # we used letter a and w 
plt.show()
```

```python
# draw plot with 2 figure-plot as one figure
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(10,5)) # 10x5 size
axe[0].boxplot(data1, labels=label1[0]) # first col
axe[1].boxplot(data2, labels=label2[0]) # second col
plt.show()
```
