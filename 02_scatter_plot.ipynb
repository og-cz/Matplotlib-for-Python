# Scatter Plot
A scatter plot is used to plot data points on a figure based on the horizontal and vertical axes. Scatter plots can be used to show the relationship between two variables. They can also show how data clusters in a dataset. One of the most striking features of drawing scatter plots in Matplotlib is the ability to set different colors and sizes for individual data points by using additional variables.

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# scatter plot
fig, axe = plt.subplots()
axe.plot(data1, label1, 'o') # indicate to be a scatter plot
plt.show()
```

```python
# draw multiple scatter in a plot
fig, axe = plt.subplots()
axe.plot(data1, label1, 'o') 
axe.plot(data2, label1, 'o') # we used label1 to just show comparison
plt.show()
```

```python
# draw plot with 2 figure-plot as one figure
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(10,3)) # 10x3 size
axe[0].plot(data1, label1, 'o') # first col
axe[1].plot(data2, label2, 'o') # second col
plt.show()
```
