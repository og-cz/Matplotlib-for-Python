# Stem plot

A stem plot (a.k.a. stem-and-leaf plot or stem plot in Matplotlib) is a type of plot used to visualize discrete data especially when the data points represent individual measured values rather than a continuous line

#### Stem plots are useful when:
- Your data is discrete, not continuous
- You want to emphasize individual values
- You want a cleaner plot than a bar chart
- You want to show spikes, peaks, pulses, or impulses
- You want to avoid connecting the dots with a line (because it might give a false trend)

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# stem plot

    # x = positions (horizontal axis)
    # y = values (vertical axis)
fig, axe = plt.subplots()
axe.stem(data1, data2)
plt.show()
```

```python
fig, axe = plt.subplots()
axe.stem(data1, data2, linefmt='r:', markerfmt='*b')
plt.show()
```

```python
fig, axe = plt.subplots()

new_data1 = [10,-20,30,-50,-15,44]
new_data2 = [20,30,40,60,70,100]

axe.stem(new_data1, new_data2, linefmt='r:', markerfmt='*b', bottom=1.0)
plt.show()
```
