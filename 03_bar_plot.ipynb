# Bar plot

The basic function in Matplotlib to create a bar chart is bar(), which display values associated with categorical data. For example, bar charts are useful if we want to plot car sales in different months, the number of people in different countries, or the average salary of a job by region.

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# bar plot
fig, axe = plt.subplots()
axe.bar(data1, label1) 
plt.show()
```

```python
# draw multiple bar line in one plot
fig, axe = plt.subplots()
axe.bar(data1, label1) 
axe.bar(data2, label1) # we used label1 to just show comparison
plt.show()
```

```python
# draw plot with 2 figure-plot as one figure
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(10,3)) # 10x3 size
axe[0].bar(data1, label1) # first col
axe[1].bar(data2, label2) # second col
plt.show()
```
