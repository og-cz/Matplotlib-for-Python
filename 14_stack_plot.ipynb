# Stack plot
the pie chart is an effective way to show the proportion of individual pieces of data as they relate to the data overall. One drawback of such charts is that they only show the data of a given moment. They do not show how parts of a population change over time. We can use stackplot to show how parts of the whole change over time.

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
fig, axe = plt.subplots()
x = range(len(label1))

data3 = [data1[_] *1.2 for _ in range(len(data1))] # creating more data by mutiplying data1 by 1.2
data4 = [data2[_] *1.2 for _ in range(len(data2))] # same as this
axe.stackplot(x, data1, data2, data3, data4, labels=label1) # label for every x values
axe.legend(loc='upper left')
plt.show()
```

# Stackplot vs Pieplot
Stackplot is NOT like a pie chart

- A pie chart shows one snapshot, one total divided into percentages.
- A stackplot shows changes across x (time, categories, samples, etc).
