# Radar plot
The radar chart, is also known as web chart, spider chart, or spider web chart, is a chart that consists of a sequence of equi-angular spokes, called radii, where each spoke represents one of the variables that is impacting our data. The data length of a spoke is proportional to the magnitude of the variable for that data point in relation to the maximum magnitude of the variable across all data points.

#### Think of it like this:

- The farther out a point is → the stronger that category is
- The closer to the center → the weaker that category is

So when you look at the shape:

- Big bulge = strong area
- Small dip = weak area

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# radar plot
fig, axe = plt.subplots(subplot_kw=dict(polar=True))

import math
angles = [n / float(len(label1)) * 2 * math.pi for n in range(len(label1))] # important thing this creates the amount of rings inside the radar plot
angles += angles[:1] # important aswell, this just take your list and append its first element to the end so the shape closes back
data1 = data1 + data1[:1] # same thing

axe.plot(angles, data1) 
axe.fill(angles, data1, 'skyblue', alpha=0.4)
plt.show()
```

```python
# Adding more than one radar to a figure

fig, axe = plt.subplots(subplot_kw=dict(polar=True))

# first 
angles = [n / float(len(label1)) * 2 * math.pi for n in range(len(label1))] 
angles += angles[:1] 

# data1 = data1 + data1[:1] # we already did this previously so just comment it out
data2 = data2 + data2[:1] # two data in one radar plot

axe.plot(angles,data1)
axe.fill(angles,data1,'skyblue',alpha=0.4)

axe.plot(angles,data2)
axe.fill(angles,data2,'teal',alpha=0.4)

plt.show()
```
