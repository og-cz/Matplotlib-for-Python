# Pie Plot
Pie charts are often used to highlight the proportion of categorical data in relation to the overall value of a dataset. Pie charts are shaped like a pie, and each category that takes up a part of the pie is shaped like a slice. The larger the slice, the greater the proportion of the overall values it represents.

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# pie plot
fig, axe = plt.subplots()

combined_data = (data1+data2)
combined_labels = (label1+label2)

axe.pie(combined_data, labels=combined_labels)
fig.show()
```

## Adding explode to the pie

Sometimes, we want to focus on a specific category, such as the one with the highest or lowest value. We can use explode to emphasize any number of wedges of the pie. explode allows us to specify an offset value for each category. As we can see in the example below, at line 5, we have set the fourth category to be offset by 0.5.

```python
# pie plot
fig, axe = plt.subplots()

combined_data = (data1+data2)
combined_labels = (label1+label2)

explode = [0,0,0,0,0,0,0,0.25] # explode! which is label z in the index of labels

axe.pie(combined_data, labels=combined_labels, autopct='%.2f%%', explode=explode)
fig.show()
```

```python
# Adding more than one pie to a figure
fig, axe = plt.subplots()
axe.pie(data1, radius=1.5, wedgeprops=dict(width=0.5),
        autopct='%.2f%%', pctdistance=0.8, labels=label1, labeldistance=1.05)
# inner circle
axe.pie(data2, radius=1.5-0.5, wedgeprops=dict(width=0.5),
        autopct='%.2f%%', pctdistance=0.7, labels=label2, labeldistance=0.3)
fig.show()
```
