# Histogram Plot

A histogram shows how often numbers appear in your dataset.
- imagine this is a test score: data = [70, 72, 75, 80, 81, 85, 88, 90, 91, 92, 95]

A histogram groups them:

70–75 → 3 values

76–80 → 2 values

81–85 → 3 values

86–90 → 2 values

91–95 → 3 values


The histogram is an important graph in statistics and data analysis. It can be used to help people quickly understand the distribution of data. 

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# histogram plot

combined_data = (data1 + data2)

fig, axe = plt.subplots()
axe.hist(combined_data)
fig.show()
```

## Drawing more than one histogram in a chart

Sometimes, we need to compare the distribution of data from different data sets. Drawing multiple histograms in the same chart can help us better understand the data. The following image shows three normally distributed sets of data:

```python
# changing style of histrogram
import random
new_data = [random.randint(1,100) for _ in range(200)] # genereate 200 data from 1 to 100 inclusive 

# let us use new data cause we need alot to just show its potential but either way we can use existing one

fig, axe = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()

axe[0][0].hist(new_data, bins=30)           # same data, differnet style
axe[0][0].set_title("set bins=30")          # title

axe[0][1].hist(new_data, density=True)      
axe[0][1].set_title("normalized")       

axe[1][0].hist(new_data, color="r")
axe[1][0].set_title("set color as red")

axe[1][1].hist(new_data, histtype='step')   # same data, differnet style
axe[1][1].set_title("step")                 # title
plt.show()
```
