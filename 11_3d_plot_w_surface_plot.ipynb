# 3D plot

A 3D surface plot is a way to visualize data that depends on two inputs (X and Y) and has one output (Z).

A surface plot answers:

> “Given X and Y, what is the height Z?”

So you give:

- X → horizontal axis
- Y → depth axis
- Z → height (value)

Matplotlib draws a smooth 3D surface through these points.

It looks like a hills, waves, or a colorful 3D terrain.

Use it when:

- You have two variables and want to see how they affect a third.
- You want a smooth curved surface.
- You want a better understanding of the shape or pattern of data.

Examples:

Temperature depending on (latitude, longitude)

- Elevation map
- Machine learning loss surface
- Mathematical functions like sin(x)+cos(y)

the data will be used will be same throughout to give show how data similarly changes in some kinda of plot

```python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# surface plot
"""
explanation:
A 3D surface plot needs a whole 2D grid of values (a table).
Your data is only 1D lists, so you don’t have a full grid.
meshgrid creates the X and Y positions for that grid.
"""

data_2d = np.array([data1, data2]) # based on the explanation so we need to create a 2d

X, Y = np.meshgrid(range(len(data1)), range(len(data_2d))) # create x and y coordinates via this function: ideally we take a the length of the array which it must have a perfect range
Z = data_2d


fig = plt.figure()
axe = fig.add_subplot(111, projection='3d')
axe.plot_surface(X, Y, Z)
plt.show()
```
