# Colorbar

The colorbar maps the values of an interval to the colors of a range. The user can then distinguish the approximate range of the data by color. colorbar is a very useful and beautiful feature.

When your plot uses colors to represent numbers (like heatmaps, intensity maps, density plots, or 3D surfaces), the viewer must know:

- What does dark blue mean?
- What value does red represent?
- Which color is highest?
- Which color is lowest?

A colorbar answers that by showing something like:

```
High value ─ red
            orange
            yellow
Low value  ─ blue
```

````python
# general data used through out the plotting
data1 = [10,20,30,50]
data2 = [20,30,40,60]
label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
````

````python
# import the needed libs
import matplotlib.pyplot as plt
import matplotlib.colors
````

````python
# colorbar - continous color
fig, axe = plt.subplots(figsize=(10,2))
cmap = plt.get_cmap('viridis')

norm = matplotlib.colors.Normalize(vmin=min(data1), vmax=max(data1)) # here!

cmapper = matplotlib.cm.ScalarMappable(norm=norm, cmap=cmap)
cmapper.set_array([])
fig.colorbar(cmapper, cax=axe, orientation='horizontal', label='viridis colormap')
plt.show()
````

````python
# colorbar - choosing different color
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20,2))
cmap = plt.get_cmap('viridis')

norm = matplotlib.colors.Normalize(vmin=min(data1), vmax=max(data1)) 

cmapper = matplotlib.cm.ScalarMappable(norm=norm, cmap=plt.get_cmap('viridis')) # set color here!
cmapper.set_array([])
fig.colorbar(cmapper, cax=axes[0], orientation='horizontal', label='viridis') # set index here!

cmapper = matplotlib.cm.ScalarMappable(norm=norm, cmap=plt.get_cmap('plasma'))
cmapper.set_array([])
fig.colorbar(cmapper, cax=axes[1], orientation='horizontal', label='plasma') 

cmapper = matplotlib.cm.ScalarMappable(norm=norm, cmap=plt.get_cmap('Accent')) 
cmapper.set_array([])
fig.colorbar(cmapper, cax=axes[2], orientation='horizontal', label='viridis')

cmapper = matplotlib.cm.ScalarMappable(norm=norm, cmap=plt.get_cmap('Reds_r')) 
cmapper.set_array([])
fig.colorbar(cmapper, cax=axes[3], orientation='horizontal', label='Reds_r') 

plt.tight_layout()
plt.show()
````
