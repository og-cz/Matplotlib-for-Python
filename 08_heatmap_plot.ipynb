# Heatmap plot
heatmap is a useful chart that we can use to show the relationship between two variables. Heatmap displays a general view of numerical data; it does not extract specific data points. It is a graphical representation of data where the individual values contained in a matrix are represented as colors.

A heatmap is a visual table where colors represent numbers.

That's it.
Instead of reading huge numeric tables, a heatmap lets your eyes quickly see:
- Which values are high
- Which values are low
- Which parts are similar or different
- Patterns and clusters

Think of it like turning a spreadsheet into a color-coded picture.

```python
# new data this is different from the previous one it must have a 2d array since its like a colored table
data_2d = [
    [0.8, 1.2, 0.3, 0.9],
    [2.5, 0.1, 0.6, 1.6],
    [1.1, 1.3, 2.8, 0.5],
    [0.2, 1.2, 1.7, 2.2]
]

label1 = ['a','b','c','d']
label2 = ['w','x','y','z']
```

```python
# import the needed libs
import matplotlib.pyplot as plt
```

```python
# heatmap plot
fig, axe = plt.subplots()
axe.set_xticks(range(len(label1))) # RANGE create evenly spaced length of each labels
axe.set_yticks(range(len(label2)))

axe.set_xticklabels(label1)
axe.set_yticklabels(label2)

im = axe.imshow(data_2d)
plt.show()
```

```python
# Adding annotations to the heatmap
fig, axe = plt.subplots()
axe.set_xticks(range(len(label1))) # RANGE create evenly spaced length of each labels
axe.set_yticks(range(len(label2)))

axe.set_xticklabels(label1)
axe.set_yticklabels(label2)

im = axe.imshow(data_2d)

for i in range(len(label2)):
    for j in range(len(label1)):
        axe.text(i, j, data_2d[i][j], horizontalalignment="center", verticalalignment="center", color="w")

plt.show()
```
